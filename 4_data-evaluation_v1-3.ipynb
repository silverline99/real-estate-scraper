```python
import pandas as pd
```

```python
# Nastavení pandas aby se:

# zobrazovaly pouze čísla na jedno desetinné místo
pd.set_option('precision', 1)

# nezkracovaly hodnoty ve sloupcích
pd.set_option('display.max_colwidth',None)
```

```python
B_listings_warehouse_BEZREALITKY_EV_df = pd.read_pickle("dataframes/B_listings_warehouse_BEZREALITKY_df.pkl")
```

```python
B_listings_warehouse_BEZREALITKY_EV_df
```

```python
# Plocha column data cleaning (odstranění m2)
B_listings_warehouse_BEZREALITKY_EV_df['Plocha'] = B_listings_warehouse_BEZREALITKY_EV_df['Plocha'].map(lambda x: str(x)[:-3]).astype(int).values
```

```python
# Cena column data cleaning (odstranění Kč)
B_listings_warehouse_BEZREALITKY_EV_df['Cena'] = B_listings_warehouse_BEZREALITKY_EV_df['Cena'].map(lambda x: str(x)[:-3]).str.replace('.', '').astype(int).values
```

```python
# Vložení sloupce CenaZaMetr do stávajícího DataFrame
B_listings_warehouse_BEZREALITKY_EV_df = B_listings_warehouse_BEZREALITKY_EV_df.assign(CenaZaMetr=B_listings_warehouse_BEZREALITKY_EV_df['Cena'] / B_listings_warehouse_BEZREALITKY_EV_df['Plocha'])
```

```python
B_listings_warehouse_BEZREALITKY_EV_df = B_listings_warehouse_BEZREALITKY_EV_df.sort_values(by=['CenaZaMetr'])
```

```python
# Export to .csv file for further (manual) processing
B_listings_warehouse_BEZREALITKY_EV_df.to_csv(r'B_listings_warehouse_BEZREALITKY_EV.csv', index = False, encoding='utf-8-sig')
```
